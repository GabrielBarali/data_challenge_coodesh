```python
# Importando bibliotecas relevantes para a atividade
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
```

```python
# Abrindo o arquivo 'netflix_titles.csv' e projetando-o
df_netflix = pd.read_csv(r'C:\Users\anaba\Downloads\data-challenge-20221121-main\data-challenge-20221121-main\data\netflix_titles.csv')
df_netflix
```

```python
# Abrindo o arquivo 'amazon_prime_titles.csv' e projetando-o
df_amazon = pd.read_csv(r'C:\Users\anaba\Downloads\data-challenge-20221121-main\data-challenge-20221121-main\data\amazon_prime_titles.csv')
df_amazon
```

## Preciso unir os datasets para fazer algumas análises. Antes disso, porém, é necessário alterar os valores da coluna 'show_id', pois parecem ser exatamente os mesmos em ambos os datasets, podendo gerar sobreposição de valores. Escolhi redefinir os valores de df_amazon na coluna 'show_id'.

```python
# Fazendo um loop para gerar uma sequência de s8808 até s18475, continuação de df_netflix
sequencia_nova2 = ['s' + str(i) for i in range(8808, 18476)]

# Salvando na variável 'data' para gerar uma nova coluna posteriormente com o nome 'show_id'
data = {'show_id': sequencia_nova2}

# Criando um novo DataFrame com a coluna 'show_id' na sequência que eu preciso
df_amazon_show_id = pd.DataFrame(data)

# Visualizando últimos 100 dados
df_amazon_show_id.tail(100)
```

```python
# Dropando a coluna 'show_id' de df_amazon
df_amazon = df_amazon.drop('show_id', axis=1)

# Concatenando df_amazon e df_amazon_show_id para obter a coluna com a sequência correta
df_amazon = pd.concat([df_amazon, df_amazon_show_id], axis=1)

# Obtendo o nome da última coluna
ultima_coluna = df_amazon.columns[-1]

# Reordenando as colunas com a última coluna no início
df_amazon = df_amazon[[ultima_coluna] + [coluna for coluna in df_amazon.columns if coluna != ultima_coluna]]

# Visualizando os dados
df_amazon
```

```python
# Concatenando df_netflix e df_amazon em df_streaming
df_streaming = pd.concat([df_netflix, df_amazon], ignore_index=True)

# Visualizando df_streaming
df_streaming
```

```python
# Visualizando os tipos de dados e valores nulos de df_streaming
df_streaming.info()
```

# 1- Top 10 atores/atrizes considerando todos os dados

```python
# Para que os valores nulos não apareçam no top 10, vou retirá-los primeiro.
# Criando uma cópia de df_streaming para fazer as alterações necessárias
df_streaming_value = df_streaming.copy()

# Dropando as linhas em que os valores de 'cast' forem nulos
df_streaming_value.dropna(subset=['cast'], inplace=True)

# Visualizando se todos os valores de 'cast' estão preenchidos
df_streaming_value.info()
```

```python
# Contando os valores de cada ator/atriz
df_streaming_value['cast'].value_counts()
```

```python
# Há celulas contendo mais de um nome de ator e atriz; logo, preciso separá-los para fazer o top 10.
# Convertendo a coluna 'cast' para o tipo string (str)
df_streaming_value['cast'] = df_streaming_value['cast'].astype(str)

# Verificando os tipos de dados após a conversão
print(df_streaming_value['cast'].dtype)
```

```python
# Criando uma lista para armazenar as séries geradas em cada iteração
series_list = []

# Iterando sobre os valores da coluna 'cast' e separando os valores em ", " (vírgula e espaço entre cada ator/atriz)
for value in df_streaming_value['cast']:
    cast = pd.Series(value.split(', '))
    series_list.append(cast)

# Concatenando as séries para criar a série final 'df_cast'
df_cast = pd.concat(series_list, ignore_index=True)

# Visualizando a série 'df_cast'
df_cast
```

```python
# Contando os valores de cada ator/atriz e mostrando-os
top_cast = df_cast
top_cast = df_cast.value_counts()
top_cast
```

```python
print(top_cast.head(10))

# Vamos considerar que '1' está incorreto, não é nome de ator ou atriz, portanto vamos
#retirá-lo do dataframe.
top_cast = top_cast.drop(labels='1')
top_cast_10 = top_cast.head(10)
print()
print(f"Top 10 atores/atrizes: {top_cast_10}")
```

```python
# Para criar um dashboard posteriormente, preciso transformar esse top 10 em um DataFrame.

# Convertendo em um DataFrame
df_top_cast_10 = top_cast_10.reset_index()
df_top_cast_10.columns = ['Nome', 'Qtd de Filmes/Shows']
df_top_cast_10
```

# 2- Top 5 países produtores de conteúdos considerando todos os dados e comparando as duas plataformas

```python
# Contando os valores de países das duas plataformas juntas
df_streaming['country'].value_counts()
```

```python
# Semelhante com o que ocorreu com os atores/atrizes, vemos que há células contendo mais de 1 país.
# Temos que separá-los.

# Criando uma cópia de df_streaming
df_streaming_country = df_streaming.copy()

# Dropando as linhas em que os valores de 'country' forem nulos
df_streaming_country.dropna(subset=['country'], inplace=True)

# Visualizando as informações e checando se os valores de 'country' estão preenchidos completamente
df_streaming_country.info()
```

```python
# Criando uma lista para armazenar as séries geradas em cada iteração
series_list_country = []

# Iterando sobre os valores da coluna 'country' e separando os valores em ", " (vírgula e espaço entre cada país)
for value in df_streaming_country['country']:
    country = pd.Series(value.split(', '))
    series_list_country.append(country)

# Concatenando as séries para criar a série final 'df_country'
df_country = pd.concat(series_list_country, ignore_index=True)

# Visualizando a série 'df_country'
print(df_country)
```

```python
# Contando os valores de cada país e mostrando-os
top_country = df_country
top_country = df_country.value_counts()
top_country_5 = top_country.head(5)
top_country_5
```

```python
# Para criar um dashboard posteriormente, preciso transformar esse top 5 países em um DataFrame.

# Convertendo em um DataFrame
df_top_countries = top_country_5.reset_index()
df_top_countries.columns = ['País', 'Qtd de Filmes/Shows']
df_top_countries
```

# 3- Mês no qual há mais adições de filmes na plataforma Netflix

```python
# Convertendo a coluna 'date' para o tipo de dados de data (datetime64)
df_netflix_month = df_netflix.copy()
df_netflix_month['date_added'] = pd.to_datetime(df_netflix_month['date_added'])

# Extraindo os meses em uma nova coluna chamada 'month'
df_netflix_month['month'] = df_netflix_month['date_added'].dt.month

# Criando uma máscara booleana para identificar as linhas em que 'type' é igual a 'TV Show'
mask = df_netflix_month['type'] == 'TV Show'

# Dropando a linha que atende à condição da máscara
df_netflix_month = df_netflix_month.drop(df_netflix_month[mask].index)

# Visualizando o mês com mais adições de filmes na Netflix
df_netflix_top_month = df_netflix_month['month'].value_counts().idxmax()
df_netflix_top_month

#Salvando a contagem em uma variável para utilizá-la no dashboard posteriormente
df_netflix_months = df_netflix_month['month'].value_counts()
df_netflix_months
```

# 4- Quantidade de filmes listados como comédia

```python
df_streaming_genre = df_streaming.copy()

# Criando uma máscara booleana para identificar as linhas em que 'type' é igual a 'TV Show'
mask = df_streaming_genre['type'] == 'TV Show'

# Dropando a linha que atende à condição da máscara
df_streaming_genre_mask = df_streaming_genre.drop(df_streaming_genre[mask].index)

# Criando uma lista para armazenar as séries geradas em cada iteração
series_list_genre = []


# Como há também células com mais de um gênero, vamos separá-los.
# Iterando sobre os valores da coluna 'listed_in' e separando os valores em ", " (vírgula e espaço entre cada gênero)
for value in df_streaming_genre_mask['listed_in']:
    genre = pd.Series(value.split(', '))
    series_list_genre.append(genre)

# Concatenando as séries para criar a série final 'df_genre'
df_genre = pd.concat(series_list_genre, ignore_index=True)

# Visualizando a série 'df_genre'
print(df_genre)

# Contando os valores de cada gênero.
df_genre.value_counts()
```

```python
#Aparentemente, há os genêros 'Comedy' e 'Comedies', ambos podem ser considerados
#filme de comédia. Temos que verificar se há outro nome próximo a esses.

# Contando a frequência de cada tipo na coluna 'genre' e salvando em genre_counts
genre_counts = df_genre.value_counts()

# Configurando o pandas para mostrar todas as linhas ao imprimir
pd.set_option('display.max_rows', None)

# Visualizando os tipos e suas contagens
print(genre_counts)
pd.set_option('display.max_rows', 20)
```

```python
#Vamos considerar como filme de comédia todos os tipos que contém 'Comedy'
#ou 'Comedies' no título. São eles: 'Comedy', 'Comedies' e 'Stand-Up Comedy'.

# Verificando se há na coluna 'listed_in' algum gênero de Comedy, Comedies ou Stand-Up Comedy
df_streaming_genre_mask['contains_comedy'] = df_streaming_genre['listed_in'].str.contains(r'\bComedy\b|\bComedies\b|\bStand-Up Comedy\b', case=False)
df_streaming_genre_mask
```

```python
# Dropando as linhas onde 'contains_comedy' é igual a False
df_streaming_genre_mask = df_streaming_genre_mask[df_streaming_genre_mask['contains_comedy']]

# Resetando o índice do DataFrame após o drop
df_streaming_genre_mask = df_streaming_genre_mask.reset_index(drop=True)

# Contando a frequência de cada valor na série
comedy_counts = df_streaming_genre_mask['contains_comedy'].value_counts()
comedy_counts

# Resultado final
qtd_comedia =comedy_counts[True]
qtd_comedia
```
